```python
import numpy as np
import matplotlib.pyplot as pt
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
data = pd.read_csv("datasets/train.csv").as_matrix()
clf = DecisionTreeClassifier()
xtrain = data[0:2000, 1:]
train_label = data[0:2000, 0]
```

```python
clf.fit(xtrain, train_label)
xtest = data[2000:, 1:]
actual_label = data[2000:,0]
```

```python
d = xtest[18]
d.shape=(28,28)
pt.imshow(255-d, cmap='gray')
print(clf.predict([xtest[18]]))
pt.show()
```

```python
clf.score(xtest, actual_label)
```
